```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("data/raw")

def load_csv(filename):
    return pd.read_csv(
        DATA_PATH / filename,
        sep=";",
        encoding="utf-8"
    )

orders = load_csv("Ir4_wc_order_stats.csv")
customers = load_csv("Ir4_wc_customer_lookup.csv")
order_items = load_csv("Ir4_woocommerce_order_items.csv")
order_itemmeta = load_csv("Ir4_woocommerce_order_itemmeta.csv")
products = load_csv("Ir4_wc_product_meta_lookup.csv")
coupons = load_csv("Ir4_wc_order_coupon_lookup.csv")
discounts = load_csv("Ir4_wdr_order_item_discounts.csv")
posts = load_csv("Ir4_posts.csv")
postmeta = load_csv("Ir4_postmeta.csv")
```

```python
datasets = {
    "Orders": orders,
    "Customers": customers,
    "Order Items": order_items,
    "Order Item Meta": order_itemmeta,
    "Products": products,
    "Coupons": coupons,
    "Discounts": discounts,
    "Posts": posts,
    "Post Meta": postmeta
}

for name, df in datasets.items():
    print(f"\n{name}")
    print("-" * 50)
    print(df.shape)
```

```python
from pathlib import Path

processed_path = Path("data/processed")
processed_path.mkdir(exist_ok=True)

for name, df in datasets.items():
    filename = name.lower().replace(" ", "_") + ".csv"
    df.to_csv(processed_path / filename, index=False)
```
