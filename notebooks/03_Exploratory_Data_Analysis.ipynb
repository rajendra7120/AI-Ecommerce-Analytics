```python
customer_summary = (
    master.groupby("customer_id")
    .agg(
        Total_Revenue=("line_total","sum"),
        Orders=("order_id","nunique"),
        Last_Order=("date_created","max")
    )
)

customer_summary.head()
```

```python
import pandas as pd

snapshot_date = master["date_created"].max() + pd.Timedelta(days=1)

customer_summary["Recency"] = (
    snapshot_date - customer_summary["Last_Order"]
).dt.days
```

```python
customer_summary.rename(columns={
    "Orders":"Frequency",
    "Total_Revenue":"Monetary"
}, inplace=True)

customer_summary.head()
```

```python
customer_summary["Customer_Type"] = "Regular"

customer_summary.loc[
    customer_summary["Monetary"] > 10000,
    "Customer_Type"
] = "VIP"

customer_summary.loc[
    customer_summary["Frequency"] == 1,
    "Customer_Type"
] = "One-Time"
```

```python
customer_summary["Customer_Type"].value_counts()
```

```python
import matplotlib.pyplot as plt

customer_summary["Customer_Type"]\
.value_counts()\
.plot(kind="bar")

plt.title("Customer Segments")

plt.show()
```

```python
#Most customers are one-time buyers.

#VIP customers contribute a disproportionately high amount of revenue.

#The company should focus on retaining VIP customers while converting one-time buyers into repeat purchasers.
```

```python
#Do 20% of products generate 80% of revenue?
```

```python
pareto = (
    master.groupby("order_item_name")["line_total"]
          .sum()
          .sort_values(ascending=False)
          .reset_index()
)

pareto.columns = ["Product","Revenue"]

pareto["Cumulative Revenue"] = pareto["Revenue"].cumsum()

pareto["Cumulative %"] = (
    pareto["Cumulative Revenue"]
    / pareto["Revenue"].sum()
) * 100

pareto.head()
```

```python
plt.figure(figsize=(15,6))

plt.plot(
    pareto["Cumulative %"].values
)

plt.axhline(80,color="red",linestyle="--")

plt.title("Pareto Analysis")

plt.ylabel("Cumulative Revenue %")

plt.show()
```

```python
#Insight
#A relatively small number of products contribute the majority of revenue.

#These products should receive priority for inventory management, advertising, and pricing optimization.
```
