```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/raw")
```

```python
def load_csv(filename):
    return pd.read_csv(
        DATA_PATH / filename,
        sep=";",
        encoding="utf-8"
    )
```

```python
orders = load_csv("Ir4_wc_order_stats.csv")
customers = load_csv("Ir4_wc_customer_lookup.csv")
order_items = load_csv("Ir4_woocommerce_order_items.csv")
order_itemmeta = load_csv("Ir4_woocommerce_order_itemmeta.csv")
products = load_csv("Ir4_wc_product_meta_lookup.csv")
coupons = load_csv("Ir4_wc_order_coupon_lookup.csv")
discounts = load_csv("Ir4_wdr_order_item_discounts.csv")
posts = load_csv("Ir4_posts.csv")
postmeta = load_csv("Ir4_postmeta.csv")
```

```python
order_itemmeta.head()
```

```python
pivot_meta = (
    order_itemmeta
    .pivot_table(
        index='order_item_id',
        columns='meta_key',
        values='meta_value',
        aggfunc='first'
    )
    .reset_index()
)

pivot_meta.head()
```

```python
order_itemmeta.head()
```

```python
pivot_meta.head()
```

```python
pivot_meta.columns.tolist()
```

```python
important_columns = [
    'order_item_id',
    '_product_id',
    '_qty',
    '_line_total',
    '_line_subtotal',
    '_line_tax',
    '_variation_id'
]

pivot_meta = pivot_meta[important_columns]

pivot_meta.head()
```

```python
pivot_meta.rename(columns={
    '_product_id': 'product_id',
    '_qty': 'quantity',
    '_line_total': 'line_total',
    '_line_subtotal': 'subtotal',
    '_line_tax': 'tax',
    '_variation_id': 'variation_id'
}, inplace=True)
```

```python
numeric_cols = [
    'product_id',
    'quantity',
    'line_total',
    'subtotal',
    'tax',
    'variation_id'
]

for col in numeric_cols:
    pivot_meta[col] = pd.to_numeric(
        pivot_meta[col],
        errors='coerce'
    )
```

```python
pivot_meta.info()
```

```python
order_details = order_items.merge(
    pivot_meta,
    on='order_item_id',
    how='left'
)

order_details.head()
```

```python
pivot_meta.info()
```

```python
order_details = order_details[
    order_details['order_item_type'] == 'line_item'
].copy()

order_details.head()
```

```python
order_details.shape
```

```python
master = order_details.merge(
    orders,
    on='order_id',
    how='left'
)

master.head()
```

```python
master.info()
```

```python
master = master.merge(
    customers,
    on='customer_id',
    how='left'
)

master.head()
```

```python
master.info()
```

```python
master = master.merge(
    products,
    on='product_id',
    how='left'
)

master.head()
```

```python
product_names = (
    posts[posts['post_type'] == 'product']
    [['ID', 'post_title']]
    .rename(columns={
        'ID': 'product_id',
        'post_title': 'product_name'
    })
)

product_names.head()
```

```python
product_names['product_id'] = pd.to_numeric(
    product_names['product_id'],
    errors='coerce'
)
```

```python
master = master.merge(
    product_names,
    on='product_id',
    how='left'
)

master.head()
```
